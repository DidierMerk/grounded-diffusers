## 1k experiments

Experiments using 1k images for training and 225 images for testing. The training images are from the train split of Pascal VOC split 1 and the testing image from its test split (so they use classes unseen by the model). For this experiment we train the segmentation model for 6 epochs using a droput rate of 0.1

FIXME: this experiment needs to be re-done because we loaded a checkpoint, damn.

```python
epochs = np.arange(6)

test_miou = np.array([
    0.3762566299079026,
    0.4063385080186331,
    0.4347923180595524,
    0.45174869272559237,
    0.4470166908568764,
    0.4740353109769278
])

train_miou = np.array([
    0.7339042354786026,
    0.7432747739988039,
    0.7517784476218359,
    0.7592911930362719,
    0.7657197331352739,
    0.7698255325436104
])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 4))

axes[0].plot(epochs, train_miou)
axes[1].plot(epochs, test_miou, "tab:orange")

axes[0].set_xlabel("Epochs")
axes[0].set_ylabel("mIoU")
axes[0].set_title("Train mIoU")
axes[0].set_yticks(np.linspace(0.7, 0.8, num=6))

axes[1].set_xlabel("Epochs")
axes[1].set_ylabel("mIoU")
axes[1].set_title("Test (unseen) mIoU")
axes[1].set_yticks(np.linspace(0.35, 0.5, num=7))
```
