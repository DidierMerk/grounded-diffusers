## 1k experiments

Experiments using 1k images for training, ~200 for validation and 225 images for testing. The training images are from the train split of Pascal VOC split 1 and the testing image from its test split (so they use classes unseen by the model). For this experiment we train the segmentation model for 6 epochs using a droput rate of 0.1.

The final test accuracy (on unseen classes) was 0.47564

```python
epochs = np.arange(10) + 1

train_miou = np.array([
    0.6280659437179565,
    0.6799320578575134,
    0.7043779492378235,
    0.7078993916511536,
    0.7125622630119324,
    0.7289522290229797,
    0.7400665283203125,
    0.7376999855041504,
    0.7358109354972839,
    0.758714534475428
])

validation_miou = np.array([
    0.6209776997566223,
    0.6578893661499023,
    0.6727970838546753,
    0.6661465764045715,
    0.6711447834968567,
    0.6713051199913025,
    0.6857122778892517,
    0.6711781024932861,
    0.6661987900733948,
    0.6899079214498613
])

test = 0.47564283316151895
```

```python
fig, ax = plt.subplots(1, 1, figsize=(14, 6))

ax.plot(epochs, train_miou)
ax.plot(epochs, validation_miou, "tab:orange")

ax.set_xlabel("Epochs")
ax.set_ylabel("mIoU")
ax.set_title("Standard Segmentation Module, 0.1 Dropout")
ax.set_xticks(np.arange(10) + 1)

ax.legend(["Train", "Validation"])
```
